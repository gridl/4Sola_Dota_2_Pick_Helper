```python
from bs4 import BeautifulSoup as BS
import requests

site_dotabuff = 'https://dotabuff.com/heroes/meta'

page_main_dotabuff = requests.get(site_dotabuff, headers = {'User-agent': 'Andy, just Andy'})
page_main_dotabuff.status_code
```

```python
page_content = BS(page_main_dotabuff.content, 'html.parser')
all_tr = page_content.find_all('tr', class_='')
all_tr
```
